```python
import pandas as pd
import numpy as np
import pydot
from sklearn.datasets import load_iris

```

#read iris.csv

```python
data=load_iris()
data.data.shape
```

x,input,target,o/p

```python
x=data.data
y=data.target
```


from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

import dicision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
print("Train data accuracy:",accuracy_score(y_true =y_train,y_pred=classifier.predict(x_train)))
print("Train data accuracy:",accuracy_score(y_true=y_test,y_pred=y_pred))
```

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus
dot_data=StringIO()
export_graphviz(classifier,out_file=dot_data,filled=True,rounded=True,feature_names=data.feature_names,class_names=data.target_names,special_characters=True)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
